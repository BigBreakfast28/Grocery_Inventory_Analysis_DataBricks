### Delta Groceries Table is created

```sql
CREATE TABLE groceries_delta
USING DELTA
AS SELECT * FROM grocery_inventory.default.grocery_inventory_new_v_1;
```

```sql
select * from groceries_delta;
```

### Update the data

```sql
update 
groceries_delta
set 
Unit_Price=0
where 
Product_Name='Arabica Coffee';
```

Update all products with "Arabica Coffee" to have a Sales_Volume of 34

```sql
update 
groceries_delta
set 
Sales_Volume = 34
where
Product_Name='Arabica Coffee';

```

```sql
select * from groceries_delta
where Product_Name='Arabica Coffee';
```

# View Total Sales By Category

```sql
SELECT Catagory, SUM(Sales_Volume) AS total_sales
FROM groceries_delta
GROUP BY Catagory
ORDER BY total_sales DESC;
```

# 

```sql
select Product_Name, SUM(Sales_Volume) AS total_sales
FROM groceries_delta
GROUP BY Product_Name
ORDER BY total_sales ASC;
```

```sql
select Product_Name as Product, Reorder_Level as Min_Quantity,Stock_Quantity
from groceries_delta
where Reorder_Level > Stock_Quantity
order by (Reorder_Level - Stock_Quantity) desc;
```

```sql
select Product_Name as Product,Date_Received, Expiration_Date, Stock_Quantity, Reorder_Level as Min_Quantity
from groceries_delta
order by Min_Quantity > Stock_Quantity desc
limit 10;
```

```sql
SELECT Product_Name, Stock_Quantity, Expiration_Date
FROM groceries_delta
ORDER BY Expiration_Date ASC, Stock_Quantity DESC;
```

```sql
SELECT 
  Product_Name AS Product,
  Date_Received,
  Expiration_Date,
  Stock_Quantity,
  Reorder_Level AS Min_Quantity,
  CASE 
    WHEN Stock_Quantity < Reorder_Level THEN 'REORDER'
    ELSE 'OK'
  END AS Inventory_Status
FROM groceries_delta
ORDER BY Inventory_Status DESC
limit 5;
```
