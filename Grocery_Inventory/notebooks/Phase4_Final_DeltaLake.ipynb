-- DELTA OPERATION #5: Time Travel Query
--

```python
%sql
-- Compare current inventory to Version 0 (initial state)

-- Aggregate comparison
SELECT
  'Current' AS version,
  COUNT(*) AS product_count,
  SUM(Stock_Quantity) AS total_stock,
  SUM(Stock_Quantity * Unit_Price) AS total_inventory_value
FROM grocery_inventory_adjusted

UNION ALL

SELECT
  'Version 0' AS version,
  COUNT(*) AS product_count,
  SUM(Stock_Quantity) AS total_stock,
  SUM(Stock_Quantity * Unit_Price) AS total_inventory_value
FROM grocery_inventory_adjusted VERSION AS OF 0;
```

```python
%sql
-- Compare specific products
WITH current_data AS (
  SELECT
    Product_Name,
    Stock_Quantity AS current_stock
  FROM grocery_inventory_adjusted
  WHERE Status = 'Active'
  ORDER BY Stock_Quantity DESC
  LIMIT 5
),
initial_data AS (
  SELECT
    Product_Name,
    Stock_Quantity AS initial_stock
  FROM grocery_inventory_adjusted VERSION AS OF 0
)
SELECT
  c.Product_Name,
  c.current_stock,
  i.initial_stock,
  (c.current_stock - i.initial_stock) AS stock_change
FROM current_data c
LEFT JOIN initial_data i ON c.Product_Name = i.Product_Name;
```
